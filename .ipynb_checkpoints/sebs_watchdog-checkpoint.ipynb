```python
# Import required libraries
import os
import time
import threading
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the folders to be monitored
folder1 = r"C:\Users\kalos\Documents\PMoPA_files\DAQ\240326\test\FluorescenceCam"
folder2 = r"C:\Users\kalos\Documents\PMoPA_files\DAQ\240326\test\ExitMode"
folder3 = r"C:\Users\kalos\Documents\PMoPA_files\DAQ\240326\test\Interferometry1Green"
folder4 = r"C:\Users\kalos\Documents\PMoPA_files\DAQ\240326\test\Interferometry2Red"
folders = [folder1, folder2, folder3, folder4]
folders = [folder1,folder2]

# Function to load a 12-bit TIFF image using OpenCV
def load_tiff_image(file_path):
    image = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
    if image is None:
        raise ValueError(f"Image at {file_path} could not be loaded")
    if image.dtype != np.uint16:
        raise ValueError("Expected 12-bit TIFF images with uint16 data type")
    # Normalize the 12-bit image data to 8-bit for display purposes
    image = (image / 4095.0 * 255).astype(np.uint8)
    return image

# Function to plot images
def plot_images(images):
    fig, axs = plt.subplots(1, len(images), figsize=(15, 5))
    for ax, img in zip(axs, images):
        ax.imshow(img, cmap='gray')
        ax.axis('off')
    plt.show(block=False)
    plt.pause(1)
    plt.close()

# Custom event handler to handle new file creations and update latest images
class NewFileHandler(FileSystemEventHandler):
    def __init__(self, folder, latest_images, lock):
        self.folder = folder
        self.latest_images = latest_images
        self.lock = lock

    def on_created(self, event):
        if event.is_directory:
            return
        with self.lock:
            if event.src_path.endswith(".tif"):
                # Update latest images list with new image path
                self.latest_images[self.folder] = event.src_path
                # Sort images by timestamp (modification time)
                sorted_images = sorted(self.latest_images.items(), key=lambda x: os.path.getmtime(x[1]), reverse=True)
                # Extract the latest image paths
                latest_paths = [path for folder, path in sorted_images[:2]]  # Get latest 2 images per folder
                # Update latest_images dictionary with sorted paths
                self.latest_images.update({folder: path for folder, path in sorted_images[:2]})

# Function to start monitoring a folder
def monitor_folder(folder, latest_images, lock):
    event_handler = NewFileHandler(folder, latest_images, lock)
    observer = Observer()
    observer.schedule(event_handler, folder, recursive=False)
    observer.start()
    try:
        while True:
            time.sleep(1)
    except KeyboardInterrupt:
        observer.stop()
    observer.join()

# Initialize latest images and locks for each folder
latest_images = {folder: None for folder in folders}
locks = {folder: threading.Lock() for folder in folders}

# Start folder monitoring in separate threads
threads = []
for folder in folders:
    thread = threading.Thread(target=monitor_folder, args=(folder, latest_images, locks[folder]))
    thread.start()
    threads.append(thread)

# Load and plot existing images initially
try:
    all_images = []
    for folder in folders:
        # Get existing images in each folder
        folder_files = sorted([os.path.join(folder, f) for f in os.listdir(folder) if f.endswith('.tif')],
                              key=os.path.getmtime, reverse=True)
        if len(folder_files) > 0:
            # Load the latest image from each folder
            latest_image = load_tiff_image(folder_files[0])
            all_images.append(latest_image)
    if all_images:
        plot_images(all_images)
except Exception as e:
    print(f"Error loading existing images: {str(e)}")

# Continuously update the plot with the latest images
try:
    while True:
        all_images = []
        for folder in folders:
            with locks[folder]:
                if latest_images[folder] is not None:
                    try:
                        image = load_tiff_image(latest_images[folder])
                        all_images.append(image)
                    except Exception as e:
                        print(f"Error loading image {latest_images[folder]}: {str(e)}")
        if all_images:
            plot_images(all_images)
        else:
            print("No new images available. Displaying latest images according to timestamp.")
        time.sleep(5)  # Adjust the sleep time as needed for refresh rate
except KeyboardInterrupt:
    for thread in threads:
        thread.join()
```
